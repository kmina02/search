## 웹 스크래핑으로 상세검색 페이지 불러오기
검색 결과 페이지의 HTML 가져오기: requests.get 함수

```python
query = '서울대입구'
url = "https://www.seoulouba.co.kr/campaign/?q={}"
r = requests.get(url.format(query))
soup = BeautifulSoup(r.text, 'html.parser')
print(soup)
```

원하는 정보가 들어있는 태그를 리스트로 가져오기: select() 메서드

```python
#검색 결과에서 특정 문구를 포함한 모든 요소 찾기
li_tags = soup.select('li.campaign_content')
```

## 체험단 리스트 출력
1. 마감된 체험단은 제외 - 완료
2. 체험단 이름, 제공내역, 링크 출력 - 완료
3. 위 정보를 담은 데이터프레임 만들기 - 완료
4. 채널이 '블로그'인 것만 선택

```python
def get_span(li_tags):
  global list, data
  list = []
  for i in range(len(li_tags)):
    # 1) campaign name
    cpn_name = li_tags[i].select_one('.t_ttl strong').text
    # 2) campaign information
    cpn_inf = li_tags[i].find('span', attrs = {'class': "basic_blue"}).get_text()
    # 3) campaign url
    cpn_url = li_tags[i].select('a.tum_img')[0]['href']
    # 4) d-day
    d_day = li_tags[i].find('div', attrs = {'class': "d_day"}).find('span').text
    # 5) cpn_list
    cpn_list = ['서울오빠', query, cpn_name, cpn_inf, cpn_url, d_day] # create a list that contains information about the campaign
    # 5) append cpn_list to list
    list.append(cpn_list)
  data = pd.DataFrame(list, columns = ['website', 'keyword', 'name', 'information', 'link', 'd-day']) # create a dataframe
  return data
```
